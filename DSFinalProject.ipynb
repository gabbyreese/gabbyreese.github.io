We tidied the data by melting the dataframe to only show the ID, National Walking Index, City Name, and Rank. Within this dataset, each city has numerous entries as the data comes from every Census block group. Our next steps are to filter the data to only show the ten cities we are interested in. ```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df = pd.read_csv('../EPA_SmartLocationDatabase_V3_Jan_2021_Final.csv')
```

```python
df.sort_values(by=["D2A_Ranked"])
```
